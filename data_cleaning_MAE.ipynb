#Task 2 — Clean the Data

```python
df=df.dropna(subset=["selling_price"])
df.shape
```

```python
df["seats"]=df["seats"].fillna(df["seats"].mode()[0])
df["seats"].info()
```

```python
df["mileage"]=df["mileage"].str.replace('kmpl','').str.strip()
```

```python
df["mileage"]=pd.to_numeric(df["mileage"])
df.info()
```

```python
df["km_driven"]=df["km_driven"].str.replace(",","").str.replace("kms","").str.replace("km","").str.strip()
```

```python
df["km_driven"]=pd.to_numeric(df["km_driven"])
num_cols=["km_driven","mileage","engine","max_power"]
for i in num_cols:
  df[i]=df[i].fillna(df[i].mean())
df.info()
```

```python
df=df.drop(columns=["Unnamed: 0","car_name","model"])
df.info()
```

```python
df["brand"]=df["brand"].str.title().str.strip()
df["brand"].unique()
```

```python
df["seller_type"].unique()
```

```python
df["seller_type"]=df["seller_type"].str.strip().str.title()
chng={'Delaer':'Dealer','Individul':'Individual','Individuall':'Individual'}
df["seller_type"]=df["seller_type"].replace(chng)
df["seller_type"].unique()
```

```python
df["transmission_type"].unique()
```

```python
df["fuel_type"]=df["fuel_type"].str.strip().str.title()
```

```python
df["fuel_type"].unique()
```

```python
df=df.drop_duplicates()
df.info()
```

#Task 3 — Compute Baseline MAE

```python
from sklearn.metrics import mean_absolute_error
df["mae"]=df["selling_price"].mean()
test_mae=mean_absolute_error(df["selling_price"],df["mae"])
print(test_mae)
```
