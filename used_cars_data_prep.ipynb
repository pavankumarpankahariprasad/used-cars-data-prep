#Task 1 — Load, Inspect, and Clean

```python
import pandas as pd
url = "/content/used_cars_messy[1].csv"
df = pd.read_csv(url)
df.shape
```

```python
df=df.dropna(subset=["selling_price"])
```

```python
df["mileage"]=df["mileage"].str.strip().str.replace("km","").str.replace("kms","").str.replace(",","")
df["mileage"]=pd.to_numeric(df["mileage"],errors='coerce')
df["mileage"]=df["mileage"].fillna(df["mileage"].median())
df.isna().sum()
```

```python
cols=["engine","max_power"]
for i in cols:
  df[i]=df[i].fillna(df[i].median())
```

```python
df["km_driven"]=df["km_driven"].str.strip().str.replace("kms","").str.replace(",","").str.replace("km","")
df["km_driven"]=pd.to_numeric(df["km_driven"])
```

```python
df=df.drop(columns=["Unnamed: 0","car_name","model"])
df
```

```python
df["seats"]=df["seats"].fillna(df["seats"].mode()[0])
```

#Task 2 — Encode Categorical Features

```python
df["transmission_type"]=df["transmission_type"].map({"Manual":0,"Automatic":1})
df
```

```python
df=pd.get_dummies(df,columns=["fuel_type","seller_type"],drop_first=True)
df
```

#Why drop_first=True is used:
Ans: In one-hot encoding, each category is converted into a separate binary column. Using drop_first=True removes one of these columns to avoid dummy variable trap (multicollinearity), where one column can be perfectly predicted from the others. This helps models (especially linear models) perform better and avoids redundant information.

#What a row of all zeros represents:

A row with all zeros in the dummy columns represents the dropped (reference) category. Since one category is removed, any instance that belongs to that category will have 0 in all the remaining dummy columns.

#Task 3 — Split and Compute Baseline MAE

```python
x=df.drop(columns=["selling_price"])
y=df["selling_price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import numpy as np
mean_value = np.mean(y_train)
y_pred = np.full(len(y_test), mean_value)
mae = mean_absolute_error(y_test, y_pred)
print(f"Baseline MAE: ₹{round(mae)}")
```
